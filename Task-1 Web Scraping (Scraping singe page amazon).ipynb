# Web Scraping a Single Amazon Product

## Objective
Scrape detailed information from a single Amazon product page 
using BeautifulSoup and Requests.

## Data Collected
- Product Title
- Current Price & Typical Price
- Rating & Total Reviews
- Delivery Charges
- Fabric Type, Care Instructions & Origin
- Customer Reviews (with ratings)

## Libraries Used
- `requests` — to send HTTP requests to Amazon
- `BeautifulSoup` — to parse HTML and extract data
- `pandas` — to structure and save data as CSV

## Output
A single-row CSV file: `1-amazon_one_product_data.csv`

## Importing Libraries

```python
from bs4 import BeautifulSoup
import requests
import time
import datetime
import pandas as pd
```

## Extracting Details

### Basic Product Information

```python
title = soup.find(id='productTitle').get_text()
price = soup.find_all('div', class_='a-section a-spacing-none aok-align-center aok-relative apex-core-price-identifier')
price = price[0].find('span', id='apex-pricetopay-accessibility-label').get_text()
typical_price = soup.find_all('div', class_='a-section a-spacing-small aok-align-center')
typical_price = typical_price[0].find('span', class_='a-offscreen').get_text()
rating = soup.find(class_='a-icon-alt').get_text()
total_reviews = soup.find(id='acrCustomerReviewText').get_text()
delivery_charges = soup.find('span' , class_='a-size-base a-color-secondary').get_text()
```

### Fabric Details

```python
fabric_details = soup.find_all('div' , class_='a-section a-spacing-micro voyager-ns-desktop-expander-wrapper')[0]
fabric_type = fabric_details.find_all('span', class_='a-color-base')[1].get_text()
care_instructions = fabric_details.find_all('span', class_='a-color-base')[3].get_text()
origin = fabric_details.find_all('span', class_='a-color-base')[5].get_text()
```

### Customer Reviews

```python
review_details = soup.find('div',id='internationalTopReviews')
review_details = review_details.find_all('span', class_='a-list-item')

customer_reviews = {}
for review in review_details:
    customer_reviews[review.find('h5').get_text()] = review.find('span', class_='a-icon-alt').get_text()
```

### Striping Data

```python
title = title.strip()
price = price.strip()
typical_price = typical_price.strip()
rating = rating.strip()[0:3]
total_reviews = total_reviews.strip()
delivery_charges = delivery_charges.strip()
fabric_type = fabric_type.strip()
care_instructions = care_instructions.strip()
origin = origin.strip()
customer_reviews = customer_reviews
```

## Create a DataFrame

```python
ds = pd.DataFrame({'Name': [title], 'Price': [price], 'Typical_Price': [typical_price], 'Rating': [rating], 'Total_Reviews': [total_reviews], 'Delivery_Charges': [delivery_charges], 'Fabric_Type': [fabric_type], 'Care_Instructions': [care_instructions], 'Origin': [origin], 'Customer_Reviews': [customer_reviews]})
ds.to_csv('1-amazon_one_product_data.csv', index=False)
ds
```
