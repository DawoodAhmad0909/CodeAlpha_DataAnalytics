# Data Cleaning — Amazon Fitness Clothing Dataset

## Objective
Clean and transform the raw scraped Amazon dataset into a 
structured, analysis-ready format.

## Cleaning Steps Performed
1. Extracted currency (PKR/$) into a separate `Currency` column
2. Removed currency symbols and converted prices to numeric (float)
3. Cleaned `Total Reviews` by removing parentheses and commas
4. Converted `Delivery Charges` to numeric values
5. Calculated `Discount Amount` and `Discount Percent` from 
   current vs typical price
6. Standardized column names for consistency

## Key Observations
- Dataset reduced from 385 to 224 rows after removing 
  rows with missing or unparseable price data
- Some products had misaligned fabric/care/origin fields 
  due to inconsistent HTML structure across Amazon listings
- Most products are priced in PKR (Pakistani Rupees)

## Output
A clean 224-row, 13-column CSV: `5-amazon_products.csv`
Ready for exploratory data analysis and visualization.

## Importing Libraries

```python
import pandas as pd
import numpy as np
```

## Reading Data From Scraped DataSet

```python
ds = pd.read_csv('4-amazon_products.csv')
ds
```

## Removing Empty & Duplicate Rows and Handling Null Values

```python
# Safely replace only exact '{}' in Customer Reviews, then drop empty rowsds['Customer Reviews'] = ds['Customer Reviews'].astype(str).str.strip()
mask = ds['Customer Reviews'].eq('{}')
ds.loc[mask, 'Customer Reviews'] = np.nan
# drop rows where every column is NaN
ds = ds.dropna(how='all')
ds = ds.drop_duplicates(ignore_index=True)
ds.to_csv('4-amazon_products_cleaned.csv', index=False)
ds
```

## Striping Numerical Values From String Columns

```python
ds['Current Price'] = ds['Current Price'].str.split('with',expand=True)[0].str.replace(',','')
ds['Currency'] = ds['Current Price'].str.replace(r'([0-9]+.)','',regex=True)
ds['Current Price'] = ds['Current Price'].str.extract(r'([0-9]+\.[0-9]+)').astype(float)
ds['Typical Price'] = ds['Typical Price'].str.replace(',','').str.extract(r'([0-9]+\.[0-9]+)').astype(float)
ds['Rating'] = ds['Rating'].str.extract(r'([0-9]+\.[0-9])').astype(float)
ds['Total Reviews'] = ds['Total Reviews'].str.replace(',','').str.extract(r'([0-9]+)').astype(float)
ds['Delivery Charges'] = ds['Delivery Charges'].str.replace(',','').str.extract(r'([0-9]+)').astype(float)
```

## Adding Discount And Discount Percentage Columns

```python
# Compute discount amount and percent
ds['Current Price'] = pd.to_numeric(ds['Current Price'], errors='coerce')
ds['Typical Price'] = pd.to_numeric(ds['Typical Price'], errors='coerce')
ds['Discount Amount'] = ds['Typical Price'] - ds['Current Price']
ds['Discount Percent'] = (ds['Discount Amount'] / ds['Typical Price'] * 100).round(2)
ds[['Product Name','Current Price','Typical Price','Discount Amount','Discount Percent']].head()
```

## Converting Prices Based On Currency

```python
print('Total Currencies')
ds['Currency'].value_counts()
```

```python
exchange_rate = 280.0  # set current USD→PKR rate
mask = ds['Currency'].astype(str).str.contains(r'\$|USD|usd|US\$', na=False)
convert_cols = [c for c in ['Current Price', 'Typical Price', 'Discount Amount', 'Delivery Charges'] if c in ds.columns]
for c in convert_cols:
	ds.loc[mask, c] = pd.to_numeric(ds.loc[mask, c], errors='coerce') * exchange_rate
ds.loc[mask, 'Currency'] = 'PKR'
if convert_cols:
	ds[convert_cols] = ds[convert_cols].round(2)
```

## Creating DataSet

```python
ds.to_csv('5-amazon_products.csv', index=False)
ds
```
