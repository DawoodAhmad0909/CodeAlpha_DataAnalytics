# Web Scraping Multiple Amazon Products

## Objective
Scrape fitness clothing products across 7 pages of Amazon 
search results, collecting 385 products in total.

## Approach
- Looped through 7 Amazon search result pages
- Used error handling (try/except) to manage failed requests
- Checked HTTP status codes to ensure successful responses
- Parsed all pages using a single BeautifulSoup object

## Data Collected
- Product Name, Current Price, Typical Price
- Rating & Total Reviews
- Delivery Charges
- Fabric Type, Care Instructions & Origin
- Customer Reviews

## Libraries Used
- `requests` — HTTP requests with custom headers to avoid blocking
- `BeautifulSoup` — HTML parsing and data extraction
- `pandas` — data structuring and CSV export

## Output
A 385-row CSV file with raw scraped Amazon product data

## Importing Libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

## Requesting URL

```python
all_html = ''
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/148.0.0.0 Safari/537.36 Edg/148.0.0.0', 'Referer': 'https://www.amazon.com'}
for i in range(1, 8):
    if i == 1:
        url = 'https://www.amazon.com/s?k=fitness+clothing&_encoding=UTF8&content-id=amzn1.sym.3fdbc1ee-e2a3-48ef-a93a-bce40e980706&pd_rd_r=84ffe51f-5b81-4cd6-a7dd-d83474e05745&pd_rd_w=P1mLu&pd_rd_wg=5Y537&qid=1780164370&xpid=dbK5xx8hp4OI8&ref=sr_pg_1'
    else:
        url = f'https://www.amazon.com/s?k=fitness+clothing&page={i}&_encoding=UTF8&content-id=amzn1.sym.3fdbc1ee-e2a3-48ef-a93a-bce40e980706&pd_rd_r=84ffe51f-5b81-4cd6-a7dd-d83474e05745&pd_rd_w=P1mLu&pd_rd_wg=5Y537&qid=1780164370&xpid=dbK5xx8hp4OI8&ref=sr_pg_{i}'
    try:
        website = requests.get(url, headers=headers, timeout=10)
        if website.status_code == 200:
            all_html += website.text
        else:
            print('Page', i, 'returned', website.status_code)
    except Exception as e:
        print('Request error on page', i, e)
soup = BeautifulSoup(all_html, 'html.parser')
```

## Importing Products Data

### Extracting Listed Products 

```python
products_list = soup.find_all('div', role='listitem')
```

### Extracting Links Of Products

```python
product_links = []

for link in products_list:
    link_tag = link.find('a',class_='a-link-normal s-line-clamp-4 s-link-style a-text-normal')
    product_links.append(link_tag['href'] if link_tag else None)
product_links
```

### Retrieving Individual Product Pages

```python
product_pages = []
for link in product_links:
    if link:
        full_link = 'https://www.amazon.com' + link
        try:
            product_page = requests.get(full_link, headers=headers, timeout=10)
            if product_page.status_code == 200:
                product_pages.append(product_page.text)
            else:
                print('Failed to retrieve product page:', full_link, 'Status code:', product_page.status_code)
        except Exception as e:
            print('Error fetching product page:', full_link, e)
```

### Creating Soup Of Each Product Page

```python
product_soups = []
for page in product_pages:
    product_soup = BeautifulSoup(page, 'html.parser')
    product_soups.append(product_soup)
```

## Retrieving Product Details

```python
titles = []
prices = []
typical_prices = []
ratings = []
total_reviews = []
delivery_charges = []
fabric_types = []
care_instructions = []
origins = []
customer_reviews = []

for soup in product_soups:
    title_tag = soup.find(id='productTitle')
    titles.append(title_tag.get_text(strip=True) if title_tag else '')

    price_tag = soup.find('div', class_='a-section a-spacing-none aok-align-center aok-relative apex-core-price-identifier')
    price = ''
    if price_tag:
        apex = price_tag.find('span', id='apex-pricetopay-accessibility-label')
        price = apex.get_text(strip=True) if apex else ''
    prices.append(price)

    typical_tag = soup.find('div', class_='a-section a-spacing-small aok-align-center')
    typical_price = ''
    if typical_tag:
        offscreen = typical_tag.find('span', class_='a-offscreen')
        typical_price = offscreen.get_text(strip=True) if offscreen else ''
    typical_prices.append(typical_price)

    rating_tag = soup.find(class_='a-icon-alt')
    ratings.append(rating_tag.get_text(strip=True) if rating_tag else '')

    total_review_tag = soup.find(id='acrCustomerReviewText')
    total_reviews.append(total_review_tag.get_text(strip=True) if total_review_tag else '')

    delivery_tag = soup.find('span', class_='a-size-base a-color-secondary')
    delivery_charges.append(delivery_tag.get_text(strip=True) if delivery_tag else '')

    # Customer reviews: handle pages without a review section and multiple possible review containers
    customer_review = {}
    review_section = (soup.find('div', id='internationalTopReviews')
                      or soup.find('div', id='reviewsMedley')
                      or soup.find('div', id='cm-cr-dp-review-list')
                      or soup.find('div', {'data-hook':'reviews-medley-footer'}))
    if review_section:
        review_items = review_section.find_all('span', class_='a-list-item')
        for review in review_items:
            title_tag = review.find('h5') or review.find('a', class_='a-size-base a-link-normal review-title')
            stars_tag = review.find('span', class_='a-icon-alt') or review.select_one('i.a-icon-star span')
            if title_tag and stars_tag:
                customer_review[title_tag.get_text(strip=True)] = stars_tag.get_text(strip=True)
    # Always append a dict (empty if none found)
    customer_reviews.append(customer_review)

    fabric_details_blocks = soup.find_all('div', class_='a-section a-spacing-micro voyager-ns-desktop-expander-wrapper')
    if fabric_details_blocks and len(fabric_details_blocks) > 0:
        fabric_details = fabric_details_blocks[0]
        fabric_spans = fabric_details.find_all('span', class_='a-color-base')
        fabric_type = fabric_spans[1].get_text(strip=True) if len(fabric_spans) > 1 else ''
        care_instruction = fabric_spans[3].get_text(strip=True) if len(fabric_spans) > 3 else ''
        origin = fabric_spans[5].get_text(strip=True) if len(fabric_spans) > 5 else ''
    else:
        fabric_type = ''
        care_instruction = ''
        origin = ''
    fabric_types.append(fabric_type)
    care_instructions.append(care_instruction)
    origins.append(origin)
```

## Creating DataSet

```python
ds = pd.DataFrame({
    'Product Name': titles,
    'Current Price': prices,
    'Typical Price': typical_prices,
    'Rating': ratings,
    'Total Reviews': total_reviews,
    'Delivery Charges': delivery_charges,
    'Fabric Type': fabric_types,
    'Care Instructions': care_instructions,
    'Origin': origins,
    'Customer Reviews': customer_reviews
})
ds.to_csv('4-amazon_products.csv', index=False)
```
